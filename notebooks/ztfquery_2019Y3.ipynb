```python
import ztfssoquery
from pathlib import Path
```

```python
targetname = "2019 Y3"

outdir = Path("../example-data") / "".join(targetname.split())
outdir.mkdir(parents=True, exist_ok=True)
outdir
```

```python
# Run the full query and URL generation process
ztfssoquery.generate_fits_urls(
    primary_des='2019 Y3',
    start_date='2025-01-01',
    end_date='2025-10-30',
    interval_days=5,
    rh_max=10,
    vmag_max=20,
    is_cutout=True,
    cutout_size="10arcmin",
    output_dir=outdir
)
```

```python
ztfssoquery.import_fitsurl(outdir/"fits_urls.txt", output_dir=outdir)
```
